```python
# Importing Libraries
from cryptography.fernet import Fernet
import base64
import os
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.kdf.pbkdf2 import PBKDF2HMAC

# Predefined master password
MASTER_PASSWORD = "shashank"

# Creating a load_salt function 
def load_salt():
    salt_file = 'salt.salt'
    # Checking if file do not exists 
    if not os.path.exists(salt_file):
        salt = os.urandom(16)
        #creating a file name 'salt_file' in 'write in btyes' mode
        with open(salt_file, mode='wb') as f:
            f.write(salt)
    # If "salt_file" already exists then 
    else:
        # Opeaning the "salt_file" in 'read in bytes' mode
        with open(salt_file, mode='rb') as f:
            salt = f.read()
    return salt

# Setting Up the Key Derivation Function (KDF)
def derive_key(master_password, salt):
    kdf = PBKDF2HMAC(
        algorithm=hashes.SHA256(),
        length=32,
        salt=salt,
        iterations=480000,
    )
    # eturning the Encryption Key
    return base64.urlsafe_b64encode(kdf.derive(master_password.encode()))

#Creating a Function to view the 'password.txt' file
def view(fer):
    try:
        with open('password.txt', mode='r') as f:
            for i in f.readlines():
                data = i.rstrip()
                user, passw = data.split("|")
                print(f"UserId/Name : {user} | Password : {fer.decrypt(passw.encode()).decode()}")
    except FileNotFoundError:
        print("Password file not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

        
#Creating a function to add new userid/name and password in 'password.txt'
def add(fer):
    username = input("Enter userid/name of the account: ")
    password = input("Enter the password: ")
    
    #Error Handling
    try:
        # Opeaning a txt file in 'a' i.e. 'Append' mode 
        with open('password.txt', mode='a') as f:
            f.write(username + " | " + fer.encrypt(password.encode()).decode() + "\n")
    except Exception as e:
        print(f"An error occurred: {e}")

# Main Function        
def main():
    entered_password = input("Enter Your Master Password: ")
    
    if entered_password != MASTER_PASSWORD:
        print("Incorrect master password. Access denied.")
        return
    
    #Loading the salt generated using load_salt() function
    salt = load_salt()
    
    #adding the master password with created salt(key) to make the salt more secure
    key = derive_key(master_password, salt)
    fer = Fernet(key)
    
    while True:
        options = input("If you want to add a new password enter ('add')\nIf you want to view the existing passwords enter ('view')\nIf you want to quit enter ('quit')\n").lower()
    
        if options == 'quit':
            break
            
        if options == 'view':
            print("=====================================")
            view(fer)
            print("=====================================")
        elif options == 'add':
            print("-------------------------------------")
            add(fer)
            print("-------------------------------------")
        else:
            print("Invalid Input")
            continue

# Directly call the main function
main()
```
